```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
df=pd.read_csv("House Data.csv")
df
```

```python
df.info()
```

```python
df.columns
```

```python
df["price"]=df["price"].str.replace(",","").str.strip()
df["price"]=pd.to_numeric(df["price"],errors="coerce")
df["GrossSquareMeters"]=pd.to_numeric(df["GrossSquareMeters"].str.extract('(\d+)')[0],errors="coerce")
df["NetSquareMeters"]=pd.to_numeric(df["NetSquareMeters"].str.extract('(\d+)')[0],errors="coerce")
df
```

```python
df.columns.tolist()
```

```python
df.info()
```

```python
df.drop('address',axis=1,inplace=True)
df.head()
```

```python
df.groupby("Category")["GrossSquareMeters"].agg(["mean","median","max","min","std","count"])
```

```python
df.groupby("NumberOfRooms")["NetSquareMeters"].agg(["mean","median","max","min","std","count"])
```

```python
df.groupby("Type")["price"].agg(["mean","median","min","max","std","count"])
```

```python
categorical = "district"
quantitative = "price"

summary_stats = df.groupby(categorical)[quantitative].agg(
    count='count',
    mean='mean',
    median='median',
    min='min',
    max='max',
    std='std'
)

summary_stats
```
