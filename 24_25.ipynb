```python
# df.isna().sum()
df=df.replace(["?"," ","-","None","nan"],pd.NA)
df.isna().sum()
```

```python
#convert area related columns to numbers
cols=["GrossSquareMeters", "BuildingAge", "NumberOfBathrooms", "NumberOfWCs",
    "NetSquareMeters", "NumberOfRooms", "FloorLocation",
    "RentalIncome", "NumberOfBalconies", "HallSquareMeters",
    "WCSquareMeters", "BathroomSquareMeters", "BalconySquareMeters"]
for col in cols:
    df[col]=pd.to_numeric(df[col].str.replace(r"[^0-9.]","",regex=True),errors="coerce")

```

```python
df["price"]=pd.to_numeric(df["price"].str.replace(r"[^0-9.]","",regex=True),errors="coerce")    
```

```python
df.info()
```

```python
#Replace the numeircal column's NaN either with Median , Mean
df["price"]=df["price"].fillna(df["price"].mean())
df.isna().sum()
```

```python
df.dtypes
```

```python
#Replace the Categorical Columns with Mode
catwalecol=df.select_dtypes(include="object").columns
catwalecol

for col in catwalecol:
    df[col]=df[col].fillna(df[col].mode()[0])

df.isna().sum()    
```

```python
#Data Transformation 
df["PricePerSqMtr"]=df["price"]/df["NetSquareMeters"]
df["PricePerSqMtr"]
```

```python
#Normalize Numeric Features 
from sklearn.preprocessing import MinMaxScaler
sc=MinMaxScaler()
df[["pricenorm","GrossSquareMetersNorm"]]=sc.fit_transform(df[["price","GrossSquareMeters"]])
```

```python
df.to_csv("newcleaneddataset.csv",index=False)
```
