```python
import numpy as np
import pandas as pd

points = {
    'P1': [0.1, 0.6],
    'P2': [0.15, 0.71],
    'P3': [0.08, 0.9],
    'P4': [0.16, 0.85],
    'P5': [0.2, 0.3],
    'P6': [0.25, 0.5],
    'P7': [0.24, 0.1],
    'P8': [0.3, 0.2],
}

# create df with point column (do this once)
df = pd.DataFrame.from_dict(points, orient='index', columns=['X','Y']).reset_index().rename(columns={'index':'point'})

# initial centroids
m1 = np.array(points['P1'])
m2 = np.array(points['P8'])

# compute distances and cluster IN-PLACE
pts = df[['X','Y']].values
df['d1'] = np.linalg.norm(pts - m1, axis=1)
df['d2'] = np.linalg.norm(pts - m2, axis=1)
df['cluster'] = np.where(df['d1'] <= df['d2'], 1, 2)

# now df contains point, X, Y, d1, d2, cluster
print(df)
```

```python
#Which cluster does P6 belong to?
print(df[df["point"]=="P6"]["cluster"].values[0])
```

```python
#What is the population around m2
# df[df["cluster"]==2].count()[0]
m2clus=df[df["cluster"]==2]
len(m2clus)
```
